```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv', sep=';')
df.head()
```

## select a few interesting features and remove nulls

```python
df2 = df[[
    'CITY', 
    'STATE', 
    'CAPITAL', 
    'IBGE_RES_POP', 
    'IBGE_RES_POP_BRAS', 
    'IBGE_RES_POP_ESTR',
    'GDP',
    'GDP_CAPITA',
    'IDHM'
]]

null_ix = [i for i in df2.index if np.isnan(df2.loc[i, 'IBGE_RES_POP'])
                                or np.isnan(df2.loc[i, 'IBGE_RES_POP_BRAS'])
                                or np.isnan(df2.loc[i, 'IBGE_RES_POP_ESTR'])
                                or np.isnan(df2.loc[i, 'GDP'])
                                or np.isnan(df2.loc[i, 'GDP_CAPITA'])
                                or np.isnan(df2.loc[i, 'IDHM'])
          ]

df2 = df2.loc[[i for i in df2.index if i not in null_ix]].reset_index(drop=True)
df2.loc[null_ix]
```

## some charts

```python
df2.sort_values(by='IBGE_RES_POP', ascending=False).head(10).sort_values(by='IBGE_RES_POP').plot.barh(
    x='CITY', y='IBGE_RES_POP', figsize=(12,7), title='20 Largest Cities in Brazil')
```

```python
cities200k = df2[df2['IBGE_RES_POP']>=200000].sort_values(by='IDHM', ascending=False).reset_index(drop=True).rename(
    columns={'IDHM':'HDI'}
)
cities200k['Average HDI'] = np.mean(cities200k['HDI'])
cities200k['vs Average'] = (cities200k['HDI'] - cities200k['Average HDI']).apply(
    lambda x: ('+' if x>=0 else '')+str(round(x, 3)))

print("Most Developed Cities in Brazil (with Population over 200,000)")
cities200k.sort_values(by='HDI', ascending=False).head(10)[['CITY', 'IBGE_RES_POP', 'STATE', 'HDI', 'vs Average']]
```

```python
print("Least Developed Cities in Brazil (with Population over 200,000)")
cities200k.sort_values(by='HDI', ascending=True).head(10)[['CITY', 'IBGE_RES_POP', 'STATE', 'HDI', 'vs Average']]
```

```python
df3 = df2.copy()
df3['IBGE_RES_POP_ESTR'] = df3['IBGE_RES_POP_ESTR'] / df3['IBGE_RES_POP']
df3['IBGE_RES_POP_LOG'] = df3['IBGE_RES_POP'].apply(math.log)

df3.plot.scatter(x='IBGE_RES_POP_LOG', y='IDHM', figsize=(12,7), title='HDI vs Log Population')
```

```python
df3[(df3['IBGE_RES_POP']>100000)].plot.scatter(
    x='IBGE_RES_POP_ESTR', y='IDHM', figsize=(12,7), 
    title='HDI vs % Foreign Population (total population over 100,000)')

df3[(df3['IBGE_RES_POP']>100000) & (df3['IBGE_RES_POP_ESTR']<=0.005)].plot.scatter(
    x='IBGE_RES_POP_ESTR', y='IDHM', figsize=(12,7), 
    title='HDI vs % Foreign Population (total population over 100,000) - ZOOMED IN')
```

```python
df3[(df3['IBGE_RES_POP']>100000)].plot.scatter(
    x='GDP_CAPITA', y='IDHM', figsize=(12,7),
    title='HDI vs per Capita GDP (total population over 100,000)')
```

## clustering

```python
from sklearn.cluster import KMeans
import math

X = df2[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)

X['IBGE_RES_POP_BRAS'] = X['IBGE_RES_POP_BRAS'] / X['IBGE_RES_POP']
X['IBGE_RES_POP_ESTR'] = X['IBGE_RES_POP_ESTR'] / X['IBGE_RES_POP']
```

```python
X_base = df2[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)

km_base = KMeans(random_state=0, n_clusters=4).fit(X_base)
X['km_base'] = km_base.labels_
X.groupby('km_base').mean()
```

```python
X_norm1 = X[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)
X_norm1['IBGE_RES_POP_BRAS'] = X_norm1['IBGE_RES_POP_BRAS'] * 10
X_norm1['IBGE_RES_POP_ESTR'] = X_norm1['IBGE_RES_POP_ESTR'] * 20
X_norm1['IBGE_RES_POP'] = X_norm1['IBGE_RES_POP'].apply(math.log)
X_norm1['IDHM'] = X_norm1['IDHM'] * 10
X_norm1['GDP_CAPITA'] = X_norm1['GDP_CAPITA'] / 10000

print(pd.DataFrame(
    [(col, min(X_norm1[col]), max(X_norm1[col]), np.mean(X_norm1[col])) for col in X_norm1.columns], columns=[
    'column', 'min', 'max', 'mean'
]))

km_norm1 = KMeans(random_state=0, n_clusters=4).fit(X_norm1)
X['km_norm1'] = km_norm1.labels_
X.groupby('km_norm1').mean()[[col for col in X.columns if 'km_' not in col]]
```

```python
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler

print('MinMaxScaler')
X_norm2 = X[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)
X_norm2 = pd.DataFrame(MinMaxScaler().fit_transform(X_norm2), columns=X_norm2.columns)
print(pd.DataFrame(
    [(col, min(X_norm2[col]), max(X_norm2[col]), np.mean(X_norm2[col])) for col in X_norm2.columns], columns=[
    'column', 'min', 'max', 'mean'
]))

print('\nMaxAbsScaler')
X_norm3 = X[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)
X_norm3 = pd.DataFrame(MaxAbsScaler().fit_transform(X_norm3), columns=X_norm3.columns)
print(pd.DataFrame(
    [(col, min(X_norm3[col]), max(X_norm3[col]), np.mean(X_norm3[col])) for col in X_norm3.columns], columns=[
    'column', 'min', 'max', 'mean'
]))

print('\nStandardScaler')
X_norm4 = X[
    ['CAPITAL', 'IBGE_RES_POP', 'IBGE_RES_POP_BRAS', 'IBGE_RES_POP_ESTR', 'GDP_CAPITA', 'IDHM']
].reset_index(drop=True)
X_norm4 = pd.DataFrame(StandardScaler().fit_transform(X_norm4), columns=X_norm4.columns)
print(pd.DataFrame(
    [(col, min(X_norm4[col]), max(X_norm4[col]), np.mean(X_norm4[col])) for col in X_norm4.columns], columns=[
    'column', 'min', 'max', 'mean'
]))
```

```python
km_norm2 = KMeans(random_state=0, n_clusters=4).fit(X_norm2)
X['km_norm2'] = km_norm2.labels_
X.groupby('km_norm2').mean()[[col for col in X.columns if 'km_' not in col]]
```

```python
km_norm3 = KMeans(random_state=0, n_clusters=4).fit(X_norm3)
X['km_norm3'] = km_norm3.labels_
X.groupby('km_norm3').mean()[[col for col in X.columns if 'km_' not in col]]
```

```python
km_norm4 = KMeans(random_state=0, n_clusters=4).fit(X_norm4)
X['km_norm4'] = km_norm4.labels_
X.groupby('km_norm4').mean()[[col for col in X.columns if 'km_' not in col]]
```
